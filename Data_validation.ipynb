```python
dummy_df = pd.get_dummies(cs_df['protocol_type'],dtype=int,drop_first=True)
dummy_df1 = pd.get_dummies(cs_df['service'],dtype=int,drop_first=True)
dummy_df2 = pd.get_dummies(cs_df['flag'],dtype=int,drop_first=True)
```

```python
cs_df = pd.concat([cs_df,dummy_df,dummy_df1,dummy_df2],axis=1)
cs_df
```

```python
cs_df = cs_df.iloc[:,:]
cs_df = cs_df.drop(columns= ["num_outbound_cmds","is_host_login",'protocol_type','service','flag'],axis=1)
cs_df
```

```python
cs_df1 = cs_df.copy()
```

Byy this data correlation I get to know which columns I should take and which I cannot.SO if correlation with 0.0 to -0.0, I take this as a no correlation withmytargeted vaue i.e class(predict cyber security risk)

```python
print(cs_df1.corr(method = "spearman")['class'].sort_values(ascending=False))
# cs_df1.corr()['class'].sort_values(ascending=False)
```

```python
corr= cs_df1.corrwith(cs_df['class'])
low_corr = corr[(abs(corr) < 0.09) &  (abs(corr) > -0.09)].index
high_corr = corr[(abs(corr) > 0.09) |  (abs(corr) < -0.09)].index
print(len(low_corr))
len(high_corr)
```

```python
cs_df1 = cs_df.drop(columns= low_corr)
cs_df1
```
