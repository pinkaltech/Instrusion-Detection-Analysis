```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score,accuracy_score,confusion_matrix,mean_squared_error
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
cs_df = pd.read_csv('cs2.csv')
cs_df = cs_df.iloc[:,1:]
cs_df
```

```python
y = cs_df[['class']]
x = cs_df.drop(columns = ['class'])
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
x_train.shape,x_test.shape
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
lr_score = r2_score(y_test,y_pred)
lr_msqr = mean_squared_error(y_test,y_pred)
lr_score
```

```python
print(lr_score)
print(lr_msqr)
```

```python
dt = DecisionTreeClassifier(random_state = 0)
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
dt_score = r2_score(y_test,y_pred)
dt_msqr = mean_squared_error(y_test,y_pred)
dt_score
```

```python
print(dt_score)
print(dt_msqr)
```

```python
rfm =   RandomForestClassifier(max_depth = 10,random_state = 0,criterion='entropy')
rfm.fit(x_train,y_train)
y_pred = rfm.predict(x_test)
rfm_score = r2_score(y_test,y_pred)
rfm_msqr = mean_squared_error(y_test,y_pred)
rfm_score
```

```python
print(rfm_score)
print(rfm_msqr)
```

```python
svm =SVC(kernel = 'rbf')
svm.fit(x_train,y_train)
y_pred = svm.predict(x_test)
svm_score = r2_score(y_test,y_pred)
svm_msqr = mean_squared_error(y_test,y_pred)
svm_score
```

```python
print(svm_score)
print(svm_msqr)
```
